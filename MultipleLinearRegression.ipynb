```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('california_housing_train.csv');
```

```python
df.head()
```

```python
y=df['total_rooms']
x=df[['population','total_bedrooms']]
plt.plot(x,y)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
sklearn_model=LinearRegression().fit(x,y)
predicted_rooms=sklearn_model.predict(x);
plt.plot(x,predicted_rooms)
plt.show()
```
